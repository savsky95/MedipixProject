```python
##UPLOAD SINGLE MATRIX AND BIN *HIT-ENERGY* NUMBER OF COUNTS. DOES NOT ACCOUNT FOR CHARGE SHARING 
#ONE ACQ = 256*256 MATRIX "FRAME" 

#Notes: if polarity is hole collecting, you will have an inverted x-axis when evaluating thl

import numpy as np 
from numpy import loadtxt
import matplotlib.pyplot as plt


#UPLOAD MATRIX AND TURN INTO 65536 LENGTH VECTOR 
m = np.loadtxt('Scan1_THL433.TXT', usecols=range(256))
A = m.flatten()
print(type(A))


#plot histogram

_ = plt.hist(A, bins='auto')
plt.title("Histogram at Fixed THL")
plt.ylabel('N Counts')
plt.xlabel('E Level [arb] ')
plt.savefig('Fixed Thl')
plt.show()

_1= plt.hist(A, bins='auto')
plt.title("Zoomed Fixed THL")
plt.ylabel('N Counts')
plt.xlabel('E Level [arb] ')
plt.xlim([0,2000])
plt.ylim([0,5000])
plt.savefig('Fixed Thl, zoom')
plt.show()





```
